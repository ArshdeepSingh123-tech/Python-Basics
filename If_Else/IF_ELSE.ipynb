# If-else in Python

```python
# login program and indentation
# email -> arsh@gmail.com
# password -> 1234

email = input('enter email')
password = input('enter password')

if email == 'arsh@gmail.com' and password == '1234':
  print('Welcome')
else:
  print('Incorrect password')


```

```python
email = input('enter email -')
password = input('enter password -')

if email == 'arsh@gmail.com' and password == '1234':
  print('Welcome')
elif email == 'arsh@gmail.com' and password != '1234':
  print('Incorrect password') # tell the user
  password = input('enter password again')
  if password == '1234':    # nested if-else (if-else ke ander if-else)
    print('Finally correct')
  else:
    print('Still incorrect')
else:
  print('Incorrect password')
```

```python
# if-else examples
# 1. Find the min of 3 given numbers
# 2. Menu Driven Program
```

```python
# min of 3 number

a = int(input('first num -'))
b = int(input('second num -'))
c = int(input('third num -'))


if a<b and a<c:
  print('smallest is',a)
elif b<c:
  print('smallest is',b)
else:
  print('smallest is',c)
```

```python
# menu driven calculator
fnum = int(input('enter 1st number - '))
snum = int(input('enter 2nd number - '))

op = input('enter the operation - ')

if op == '+':
  print(fnum + snum)
elif op == '-':
  print(fnum - snum)
elif op == '*':
  print(fnum * snum)
elif op == '/':
  print(fnum / snum)
```

```python
# menu driven ATM

menu = input("""Hi! how can I help you.
1. Enter 1 for pin change
2. Enter 2 for balance check
3. Enter 3 for withdrawl
4. Enter 4 for exit
""")

if menu == '1':
  print('pin change')
elif menu == '2':
  print('balance')
else:
  print('exit')
```
