## 9. Take a year as input and determine whether it is a leap year.

```python
a = int(input('Enter a year  - '))

if a % 400 == 0:
  print('Leap Year')
elif a % 4 ==0 and a % 100 != 0:
  print('Leap Year')
else:
  print('Not a Leap Year')
```

12. Take:

first number

second number

operator (+, -, *, /)

Perform the operation using if-elif-else.

```python
fnum = int(input('Enter a first number  - '))
snum = int(input('Enter a second number  - '))
op = input('Enter a operator  - ')

if op == '+':
  print(fnum + snum)
elif op == '-':
  print(fnum - snum)
elif op == '*':
  print(fnum * snum)
elif op == '/':
  print(fnum / snum)
else:
  print('Invalid Operator')
```

15. Take purchase amount:

Above 5000 → 20% discount

Above 2000 → 10% discount

Otherwise → No discount

Print the final amount.

```python
n = int(input('Enter a amount - '))

if n > 5000:
  a = n - (n * 20/100)
  print('Final price - ',a)
elif n > 2000 and n <= 5000:
  b = n - (n * 10/100)
  print('Final price - ',b)
else:
  print('NO discount - ',n)
```

# OR

```python
amount = float(input("Enter purchase amount: "))

if amount > 5000:
    discount = amount * 20 / 100
elif amount > 2000:
    discount = amount * 10 / 100
else:
    discount = 0

final_amount = amount - discount

print("Discount =", discount)
print("Final Amount =", final_amount)
```

# 16. Check Armstrong Number (3-digit)
153 = 1³ + 5³ + 3³

```python
n = int(input('Enter a number  - '))

original = n

a = n % 10
n = n // 10

b = n % 10
n = n // 10

c = n % 10

sum = a**3 + b**3 + c**3

if sum == original:
  print('Armstrong Number')
else:
  print('Not Armstrong Number')
```

18. Grade System
