# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

### Creating Sets

```python
# empty
s = set()
print(s)
print(type(s))

# 1D and 2D
s1 = {1, 2, 3}
print(s1)
print(type(s1))

#s2 = {1, 2, 3,{4, 5}}
#print(s2)

# homo and hetro
s3 = {1,'hello',4.5,(1,2,3),True}
print(s3)

# using type conversion
s4 = set([1,2,3])
print(s4)

# duplicates not allowed
s5 = {1,1,2,2,3,3}
print(s5)

# set can't have mutable items
s6 = {1,2,[3,4]}
print(s6)
```

```python
s1 = {1,2,3}
s2 = {3,2,1}

print(s1 == s2)
```

### Accessing Items

```python
s1 = {1,2,3,4}
s1[0:3]
```

### Editing Items

```python
s1 = {1,2,3,4}
s1[0] = 100
```

### Adding Items

```python
S = {1,2,3,4}
# add
S.add(5)
print(S)

# update
S.update([6,7,8])
print(S)
```

### Deleting Items

```python
# del
s = {1,2,3,4,5}
print(s)
del s
print(s)

```

```python
s = {1,2,3,4,5}
del s[0]
print(s)
```

```python
# discard
s = {1,2,3,4,5}
s.discard(5)
print(s)

s.discard(50)
print(s)
```

```python
# remove
s = {1,2,3,4,5}
s.remove(5)
print(s)
```

```python
s = {1,2,3,4,5}
s.remove(50)
print(s)
```

```python
# s.pop()
s = {1,2,3,4,5}
s.pop()
print(s)
```

```python
# clear
s.clear()
print(s)
```

### Set Operation

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# Union(|)
s1 | s2

# Intersection(&)
s1 & s2

# Difference(-)
s1 - s2
s2 - s1

# Symmetric Difference(^)
s1 ^ s2

# Membership Test
1 in s1
1 not in s1

# Iteration
for i in s1:
  print(i)
```

### Set Functions

```python
# len/sum/min/max/sorted
s = {3,1,4,5,2,7}
len(s)
sum(s)
min(s)
max(s)
sorted(s)
sorted(s,reverse=True)
```

```python
# union/update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 | s2
s1.union(s2)
s1.update(s2)

print(s1)
print(s2)
```

```python
# intersection/intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2)

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2)

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 = {7,8,5,6}

s1.isdisjoint(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issuperset(s2)
```

```python
# copy
s1 = {1,2,3}
s2 = s1.copy()

print(s1)
print(s2)
```

### Frozenset
Frozen set is just an immutable version of a Python set object

```python
# create frozenset
fs1 = frozenset([1,2,3])
fs2 = frozenset([3,4,5])

fs1 | fs2
```

```python
# what works and what does not
# works -> all read functions
# does't work -> write operations
```

```python
# When to use
# 2D sets
fs = frozenset([1,2,frozenset([3,4])])
fs
```
