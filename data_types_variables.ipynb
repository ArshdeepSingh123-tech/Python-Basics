# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

```python
# Arithmetric Operators

print(5+6) #addition

print(5-6) #substraction

print(5*6) # multiplication

print(5/2) # division

print(5//2)  #integer division

print(5%2)   # tell remainder

print(5**2)  # power
```

```python
# Relational Operators

print(4 > 5)

print(4 < 5)

print(4 >= 4)

print(4 <= 4)

print(4 == 4)

print(4 != 4)
```

```python
# Logical Operators

print(1 and 0) # give 1 when both 1 and 1

print(1 or 0) # give 1 when both are 1 ,when 1 is 1

print(not 1) # zero ko 1 bna deta hai or 1 ko zero
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3) # 01 and 11 = 01 -> 2

# bitwise or
print(2 | 3) # 01 or 11 = 11 -> 3

# bitwise xor
print(2 ^ 3) # 01 or 11 = 10 ->1

# bitwise not
print(~3) # ∼n=−(n+1)

# bitwise right shift
print(4 >> 2) # n≫k=n/2^k

# bitwise left shift
print(5 << 2) # n≪k=n×2^k


```

```python
# Assignment Operators

 # =
 # a = 2

a = 2

a += 2 # a = a + 2

print(a)
```

```python
# Membership Operators

# in/not in

print('D' in 'Delhi')
print('D' not in 'Delhi')

print(1 in [2,3,4,5,6])

```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input('Enter a 3 digit number'))

# 345%10 -> 5
a=number%10

number = number//10

# 34%10 -> 4
b=number%10

number = number//10

# 3%10 -> 3
c=number%10

print(a + b + c)

```
