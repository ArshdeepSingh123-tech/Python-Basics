# Pattern Questions

1. Print:

```text
*
**
***
****
*****
```

````python
n = int(input('Enter a number - '))

for i in range(1,n+1):
  for j in range(1,i+1):
    print('*',end='')
  print()
````

2. Print:
```
1
12
123
1234
12345

````python
n = int(input('Enter a number - '))

for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end='')
  print()
````

3. Print:
```text
*****
****
***
**
*
```

````python
n = int(input('Enter a number - '))

for i in range(n,0,-1):
    for j in range(i):
        print('*',end='')
    print()
````

5. Print:
```
1
121
12321
1234321

````python
n = int(input('Enter a number - '))

for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end='')

  for j in range(i-1,0,-1):
    print(j,end='')

  print()
````
