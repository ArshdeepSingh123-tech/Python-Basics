# Logic Building Questions

1. Check whether a number is an Armstrong number (3 digits).


```python
n=int(input("Enter a number:"))

original = n
b = 0

while n > 0:
  a = n % 10
  b = b + (a**3)
  n = n // 10

if b == original:
    print("Armstrong number")
else:
    print("Not an Armstrong number")
```

2. Find the largest digit in a number.

   Example: 5832 → 8

```python
n = int(input('Enter a number -  '))

largest = 0

while n > 0:
  a = n % 10

  if a > largest:
    largest = a

  n = n // 10

print('largest digit- ',largest)
```

3. Find the smallest digit in a number.

```python
n = int(input('Enter a number -  '))

smallest = 9

while n > 0:
  a = n % 10

  if a < smallest:
    smallest = a

  n = n // 10

print('smallest digit- ',smallest)
```

4. Count the number of even digits in a number.

   Example: 123456 → 3

```python
n = int(input('Enter a number -  '))

count = 0

while n > 0:
  a = n % 10

  if a % 2 == 0:
    count = count + 1

  n = n // 10

print('Count of even digits - ',count)
```

5. Calculate:
   
   1 + 2 + 3 + ... + N


```python
n = int(input('Enter a number -  '))

sum = 0

for i in range(1,n+1):
  sum = sum + i

print('sum - ', sum)
```

6. Calculate:

```python
n = int(input('Enter a number -  '))

sum = 0

for i in range(1,n+1):
  sum = sum + i**2

print('sum - ', sum)
```

7. Print the multiplication table of a number.

```python
n = int(input('Enter a number -  '))

for i in range(1,11):
  print(n,'*',i,'=',n*i)
```

8. Check whether a number is prime.

```python
n = int(input('Enter a number -  '))

is_prime = True

for i in range(2,n):
  if n % i == 0:
    is_prime = False
    break

if is_prime:
  print('Prime number')
else:
  print('Not a prime number')
```

9. Print all prime numbers between 1 and N.

```python
n = int(input('Enter a number -  '))

for i in range(2,n+1):
  is_prime = True

  for j in range(2,i):
    if i % j == 0:
      is_prime = False
      break

  if is_prime:
    print(i)
```
