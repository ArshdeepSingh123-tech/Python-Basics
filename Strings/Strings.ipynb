## Editing and Deleting in Strings

```python
s = 'hello world'
s[0] = 'H'

# Python strings are immutable
```

```python
s = 'hello world'
del s # s string is deleted
print(s)
```

```python
s = 'hello world'
del s[-1:-5:2]
print(s) # this throw an error because string are immutable(string cannot be changed)
```
